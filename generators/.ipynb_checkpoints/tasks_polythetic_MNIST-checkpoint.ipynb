```python
import matplotlib.pyplot as plt
```

```python
class Config:
    path = 'PATH_TO_DATA_FOLDER'
    shot = 24
    query = 8
    train_way = 2
    examples_per_group = 32
    groups_per_class = 2
config = Config()
```

#### Clean polythetic MNIST

```python
from clean_polythetic_MNIST import task_generator
```

```python
g = task_generator(config)
data_shot, label_shot, data_query, label_query = g.get_shot_query(config, 'cpu', xor_task=False)

plt.figure(figsize=(16, 24))
plt.suptitle('Support set. Monothetic task')
for i in range(96):
    plt.subplot(12, 8, i+1)
    plt.imshow(data_shot[i].numpy().transpose(1, 2, 0))
    plt.axis('off')
    plt.title('Class: ' + str(label_shot[i].numpy().astype(int)));
plt.subplots_adjust(top=0.95)
```

```python
g = task_generator(config)
data_shot, label_shot, data_query, label_query = g.get_shot_query(config, 'cpu', xor_task=True)

plt.figure(figsize=(16, 24))
plt.suptitle('Support set. Polythetic task')
for i in range(96):
    plt.subplot(12, 8, i+1)
    plt.imshow(data_shot[i].numpy().transpose(1, 2, 0))
    plt.axis('off')
    plt.title('Class: ' + str(label_shot[i].numpy().astype(int)));
plt.subplots_adjust(top=0.95)
```
